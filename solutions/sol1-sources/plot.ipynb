```python
import pandas as pd

df = pd.read_csv('train_log.csv')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.subplots_adjust(right=0.75)

twin1 = ax.twinx()

ax.plot(df['epoch'], df['train_loss'], label='Train Loss', color='red')
twin1.plot(df['epoch'], df['val_acc'] * 100.0, label='Validation Accuracy')

ax.set_ylim(0, 15)
ax.set_ylabel("Train Loss")
ax.set_xlabel("Epoch")

twin1.set_ylim(0, 100)
twin1.set_ylabel("Validation Accuracy (%)")
# legend at top left
ax.legend(loc='upper left')
twin1.legend(loc='upper right')
```
